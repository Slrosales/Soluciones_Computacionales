```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import lambdify
```

```python
def lagrange_interpolation(points):
    # Extracción de coordenadas x e y de los puntos
    x_values, y_values = zip(*points)
    x = symbols('x')

    # Inicializa el polinomio interpolador
    interpolating_polynomial = 0

    for i in range(len(points)):
        # Calcula Li(x) para cada i
        Li = prod([(x - x_values[j]) / (x_values[i] - x_values[j]) for j in range(len(points)) if j != i])

        # Agrega Li(x) * yi al polinomio interpolador
        interpolating_polynomial += y_values[i] * Li

    return interpolating_polynomial

# Ejemplo de puntos (x, y) para interpolación
points = [(0, -0.6), (0.1, -5.89483), (0.3, -5.65014), (0.6, -5.17788), (1, -4.28172)]

# Calcula el polinomio interpolador
interpolating_poly = lagrange_interpolation(points)

# Muestra el polinomio interpolador
print("Polinomio interpolador:")
print(interpolating_poly)
```

```python
x_values= [0,0.1,0.3,0.6,1]
y_values = [-0.6,-5.89483,-5.65014,-5.17788,-4.28172]
interpolating_func = lambdify('x', interpolating_poly, 'numpy')

# Crea un rango de valores de x para la gráfica
x_range = np.linspace(min(x_values), max(x_values), 400)

# Evalúa la función en el rango de valores de x
y_interpolated = interpolating_func(x_range)

# Grafica los puntos y la función interpolada
plt.scatter(x_values, y_values, label='Puntos dados')
plt.plot(x_range, y_interpolated, label='Función interpolada', color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interpolación de Lagrange')
plt.grid(True)
plt.show()
```
