```python
def coef(x, X, Y):
    n = len(X)
    result = 0.0
    for i in range(n):
        term = Y[i]
        for j in range(n):
            if i != j:
                term *= (x - X[j]) / (X[i] - X[j])
        result += term
    return result
```

```python
x = [0.0, 0.1, 0.3, 0.6, 1.0]
y = [-0.6, -5.89483, -5.65014, -5.17788, -4.28172]
```

```python
# Generar puntos a lo largo del rango de X
x_n = np.linspace(0, 1, 100)
L = [coef(x, X, Y) for x in x_n]

# Graficar los puntos y el polinomio interpolador
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, label='Puntos de datos', color='blue')
plt.plot(x_n, L, color='red')
plt.legend()
```

```python
X = np.array([0.0, 0.1, 0.3, 0.6, 1.0,1.1])
Y = np.array([-0.600000, -5.89483, -5.65014, -5.17788, -4.28172, -3.99583])

# Función para calcular el polinomio interpolador de Lagrange
def lagrange_interpolation(x, X, Y):
    n = len(X)
    result = 0.0
    for i in range(n):
        term = Y[i]
        for j in range(n):
            if i != j:
                term *= (x - X[j]) / (X[i] - X[j])
        result += term
    return result


x_range = np.linspace(0, 1.2, 100)
y_interpolated = [lagrange_interpolation(x, X, Y) for x in x_range]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, label='Puntos de datos', color='red')
plt.plot(x_range, y_interpolated, label='Polinomio Interpolador de Lagrange', color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Interpolación de Lagrange')
plt.grid(True)
plt.show()
```
